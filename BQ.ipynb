```python
import sys
import pandas as pd
from google.cloud import bigquery
```

```python
client = bigquery.Client("flink-data-prod")
transactions_2018_query = f"""
  select from_address, to_address from bigquery-public-data.crypto_ethereum.transactions where 
(DATE(block_timestamp)>= DATE("2018-01-01") and DATE(block_timestamp) < DATE("2018-02-01"))
or (DATE(block_timestamp)>= DATE("2018-01-04") and DATE(block_timestamp) < DATE("2018-02-04"))
or (DATE(block_timestamp)>= DATE("2018-01-07") and DATE(block_timestamp) < DATE("2018-01-07"))
or (DATE(block_timestamp)>= DATE("2018-01-10") and DATE(block_timestamp) < DATE("2018-02-10"))
    """
transactions_2018_query_df = client.query(transactions_2018_query).to_dataframe()

transactions_2018_query_df.to_csv("transactions_2018_query_df.csv")
```
